# **Box-MÃ¼ller Transform**

$$
U1,U2 \implies U(0,1)
$$

```python
u = np.random.rand(1000)
```

```python
plt.hist(u)
```

```python
u_1 = u[:500]
u_2 = u[500:]
```

```python
z_1 = np.sqrt(-2*np.log(u_1))*np.cos(2*np.pi*u_2)
z_2 = np.sqrt(-2*np.log(u_1))*np.sin(2*np.pi*u_2)
```

```python
z = np.concatenate((z_1,z_2),axis=0)
z
```

```python
plt.subplot(1,2,1)
plt.hist(z)
plt.subplot(1,2,2)
plt.hist(z,bins=100)
```

```python
np.var(z),np.std(z),np.mean(z)
```

# **Sample to General Normal**

$$
N-(\mu,\sigma)
$$

$$
Y=\mu + \sigma z
$$

```python
mean = 3
sigma = 2
Y = mean + sigma*z
```

```python
plt.hist(z,bins=100)
plt.hist(Y,bins=100)
```
