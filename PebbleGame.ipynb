```python
def Error(nums):
  return np.std(nums)/np.sqrt(len(nums))
```

**Direct Sampling**

```python
import random
import numpy as np

def direct_pi(N):
    n_hits = 0
    for i in range(N):
        x, y = random.uniform(-1.0, 1.0), random.uniform(-1.0, 1.0)
        if x ** 2 + y ** 2 < 1.0:
            n_hits += 1
    return n_hits
nums = []
n_runs = 1600
n_trials = 10000
for run in range(n_runs):
    nums.append(4.0 * direct_pi(n_trials) / float(n_trials))
np.mean(nums),np.std(nums),Error(nums)
```

**Markov_pi**

```python
import random
import numpy as np
def markov_pi(N, delta):
    x, y = 1.0, 1.0
    n_hits = 0
    for i in range(N):
        del_x, del_y = random.uniform(-delta, delta), random.uniform(-delta, delta)
        if abs(x + del_x) < 1.0 and abs(y + del_y) < 1.0:
            x, y = x + del_x, y + del_y
        if x**2 + y**2 < 1.0: n_hits += 1
    return n_hits

nums_2 = []
n_bunch = 100
n_runs = 100
n_trials = 4000
delta = 0.1
delta_2 = 0.5
for bunch in range(n_bunch):
  nums = []
  for run in range(n_runs):
      nums.append(4.0 * markov_pi(n_trials, delta) / float(n_trials))
      #nums_2.append(4.0 * markov_pi(n_trials, delta_2) / float(n_trials))
  nums_2.append(np.mean(nums))


np.std(nums_2),np.std(nums),Error(nums)

```
