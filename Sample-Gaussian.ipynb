#**Gaussian Integral**

$$
\int _ {-\infty}^{\infty} e^{\frac{-1}{2}x^{2}}\sqrt\frac{1}{2\pi}\;dx = I
$$

$$
\int _ {-\infty}^{\infty} \sqrt\frac{1}{2\pi} \;dx\int _ {-\infty}^{\infty} \sqrt\frac{1}{2\pi}e^{\frac{-1}{2}x^{2}\frac{-1}{2}y^{2}} \;dy = I^{2}
$$

**change of varibles**
$$
x = r\cos(\theta)
$$

$$
y = r\sin(\theta)
$$

$$
\;dx\;dy = r\;d\theta\;dr
$$

$$
\int _ 0^\infty \int_0 ^ {2\pi} e^{-r^{2}} r\;d\theta\;dr,
$$

$$
e^{-Ψ} = Υ
$$

$$
\int _ {0}^{2\pi} \frac{1}{2\pi} \;d\theta\int _ {0}^{1}\;dΥ
$$


1. $\theta$ is a random number between 0 and $2\pi$.
2. $Υ$ is a random number between 0 and 1.

```python
import random, math
import numpy as np
import matplotlib.pyplot as plt


def gauss_test(sigma):
    phi = random.uniform(0.0, 2.0 * math.pi)
    Upsilon = random.uniform(0.0, 1.0)
    Psi = - math.log(Upsilon)
    r = sigma * math.sqrt(2.0 * Psi)
    x = r * math.cos(phi)
    y = r * math.sin(phi)
    return [x, y]


samples = [[],[]]
nsamples = 10000
for sample in range(nsamples):
  [x, y] = gauss_test(1.0)
  samples[0].append(x)
  samples[1].append(y)

plt.hist2d(samples[0],samples[1])
plt.show()
```

```python
np.mean(samples[0]),np.std(samples[0]),np.mean(samples[1]),np.std(samples[1])
```

```python
plt.plot(samples[0],samples[1],".")
plt.show()
```

```python
plt.hist(samples[0])
plt.hist(samples[1])
plt.show()
```

# **Gaussian in 2-D**

```python
import random
samples = [[],[]]
nsamples = 10000
for sample in range(nsamples):
  x, y = random.gauss(0.0,0.1), random.gauss(0.0,0.1)
  samples[0].append(x)
  samples[1].append(y)

plt.hist2d(samples[0],samples[1])
plt.show()
```

```python
plt.plot(samples[0],samples[1],".")
plt.show()
```

# **Gaussian in 3-D**

```python
import matplotlib.pyplot as plt
import random
samples = [[],[],[]]
nsamples = 10000
for sample in range(nsamples):
  x, y ,z = random.gauss(0.0,0.1), random.gauss(0.0,0.1), random.gauss(0.0,0.1)
  samples[0].append(x)
  samples[1].append(y)
  samples[2].append(z)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot3D(samples[0],samples[1] ,samples[2],".")
plt.show()
```

# **Gaussian in N-Dimensions**




**Direct Surface**

*We can change dimension how we want in the same way.
For example let's choose 3-d to plot.*  

```python
import random, math

samples = []
dimensions = 3
nsamples = 10000
for sample in range(nsamples):
    R = [random.gauss(0.0, 1.0) for d in range(dimensions)]
    radius = math.sqrt(sum(x ** 2 for x in R))
    samples.append([x / radius for x in R])

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
for i in range(len(samples)):
    ax.plot3D(samples[i][0],samples[i][1] ,samples[i][2],".")
plt.title("uniform distribution on the sphere")
plt.show()
```

*this sphere is isotrophic(the probability distribution is uniform on the surface)*

**Direct Sphere**

**change of variables**
$$
(x_0, \ldots, x_{d-1}) \implies (r,Ω)
$$

*in n-dimensions, there is no function depending on angular variable*

*so,  $\phi(r) = r^{d-1}dr$ (0 < r < 1)*

*distribution of radii inside unit sphere of constant density*


*sampled through $(random.uniform(0.0,1.0))^{1/d}$*

*let's look at 3-d*

```python
import random, math
samples = [[],[],[]]
nsamples = 10000
for sample in range(nsamples):
  x, y, z = (random.gauss(0.0, 1.0),
              random.gauss(0.0, 1.0),
               random.gauss(0.0, 1.0))
  length = random.uniform(0.0, 1.0) ** (1.0 / 3.0) \
                    / math.sqrt(x ** 2 + y ** 2 + z ** 2)
  samples[0].append(x * length)
  samples[1].append(y * length)
  samples[2].append(z * length)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot3D(samples[0],samples[1] ,samples[2],".")
plt.show()

```
